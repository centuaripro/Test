```python
import pandas as pd
```

You can create a pandas dataframe from a dictionary as seen below

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Boston', 'Chicago', 'Denver', 'Seattle'],
    'Score': [85, 92, 78, 95, 89]
}

df = pd.DataFrame(data)
print(df)
```
