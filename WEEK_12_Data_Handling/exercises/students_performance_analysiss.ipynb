```python
import pandas as pd

from pprint import pprint
```

```python
def load_student_records(filename):
    df = pd.read_csv(filename)

    subject_cols = [ col for col in df.columns.to_list() if "_score" in col]
    
    for col in subject_cols:
        # fill missing values with either the mean of all values in the respective columns or the minimum value present in the column
        # df[col].fillna(df[col].mean() inplace=True)

        df.fillna({col: df[col].min()}, inplace= True)

        # for xample when doing 'df[col].method(value, inplace =True)',
        # try using 'df.method({col: value}, inplace=True)' instead,
        # to perform the operation inplace on the original object,
        # or try to avoid an inplace operation using 'df[col] = df[col].method(value)'.

        # Convert all numeric values to their various data type of either float or integer

        df[col] = pd.to_numeric(df[col], errors= "coerce")

    pd.to_numeric(df["attendance_rate"], errors = "coerce")

    return df
```

```python
def calculate_student_average(df):

    subject_cols = [ col for col in df.columns.to_list() if "_score" in col and col != "average_score"]
    
    df["average_score"] = df[subject_cols].mean(axis =1, skipna= True)

    return df
```

```python
def get_grade_letter(avg_score):
    if avg_score > 100:
        raise Exception("Average Score cannot be greater than 100")
    
    if avg_score >= 90:
        return "A"
    elif avg_score >= 80:
        return "B"
    elif avg_score >= 70:
        return "C"
    elif avg_score >= 60:
        return "D"
    elif avg_score >= 60:
        return "E"
    else: 
        return "F"

def assign_grade_letter(df):
    df["grade_letter"] = df["average_score"].apply(get_grade_letter)

    return df
```

```python
def find_top_performers_by_subject(df):

    subject_cols = [ col for col in df.columns.to_list() if "_score" in col]

    top_performers = {}

    for subject in subject_cols:
        max_score = df[subject].max()
        top_students = df[df[subject] == max_score][["student_id", "first_name", "last_name", subject]]
        top_performers[subject] = top_students

    return top_performers
    

    
```

```python
def calc_correlation(df: pd.DataFrame):
    return df["average_score"].corr(df["attendance_rate"])
```

```python
students_df = load_student_records("../files/students_record.csv")

students_df = calculate_student_average(students_df)

students_df = assign_grade_letter(students_df)

top_oerformers= find_top_performers_by_subject(students_df)

correlation = calc_correlation(students_df)

print(correlation)

```
